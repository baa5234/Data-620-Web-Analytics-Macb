```python
#Name : Blessing Anoroh
#Assignment 1, wk2 - Data 620 - Web Analytics
#Date : Week 2, February 1, 2026


import networkx as nx
import matplotlib.pyplot as plt


names = {
    "A": "Andre","B": "Beverly","C": "Carol","D": "Diane","E": "Ed",
    "F": "Fernando","G": "Garth","H": "Heather","I": "Ike","J": "Jane"
}

edges = [
    ("A","B"), ("A","C"), ("A","D"), ("A","F"),
    ("B","D"), ("B","E"),
    ("C","D"), ("C","F"),
    ("D","E"), ("D","F"), ("D","G"),
    ("E","G"),
    ("F","G"), ("F","H"),
    ("G","H"),
    ("H","I"),
    ("I","J"),
]

G = nx.Graph()
G.add_edges_from(edges)

position = {
    "C": (-0.2,  1.2),
    "A": (-1.1,  0.6),
    "F": ( 0.7,  0.6),
    "B": (-1.1, -0.4),
    "G": ( 0.7, -0.4),
    "E": (-0.2, -1.1),
    "D": (-0.2,  0.1),
    "H": ( 1.6,  0.1),
    "I": ( 2.6,  0.1),
    "J": ( 3.6,  0.1),
}

plt.figure(figsize=(12, 6))
nx.draw(G, position, with_labels=False, node_size=1800, width=3)
nx.draw_networkx_labels(G, position, labels=names, font_size=12)
plt.axis("off")
plt.title('Assignment 1 : Krackhardt Kite â€” hello, graph world', fontsize=16)
plt.show()
```
